```python
%matplotlib inline
%matplotlib inline
import matplotlib.pyplot as plt
import lightkurve as lk
```

```python
TIC = 'TIC 278892590'
sector_data = lk.search_lightcurve(TIC)
sector_data
```

```python
x = 0
while x <= 6:
    lc = sector_data[x].download()
    lc.plot(linewidth = 0, marker = '.', color = 'blue', alpha = 1, markersize = 1)
    x += 1
```
